#### 1. Изучить линейную регрессию (теоретические основы).

Линейная регрессия (англ. Linear regression) — используемая в статистике регрессионная модель зависимости одной (объясняемой, зависимой) переменной y от другой или нескольких других переменных (факторов, регрессоров, независимых переменных) x с линейной функцией зависимости.

Регрессионная модель

$ y=f(x,b)+\varepsilon ,~E(\varepsilon )=0$

где $b$ — параметры модели, $ \varepsilon $  — случайная ошибка модели; называется линейной регрессией, если функция регрессии $f(x,b)$ имеет вид

$f(x,b)=b_{0}+b_{1}x_{1}+b_{2}x_{2}+...+b_{k}x_{k}$

где $b_{j}$ — параметры (коэффициенты) регрессии, $x_{j}$ — регрессоры (факторы модели), $k$ — количество факторов модели

#### 2. Проверить наличие мультиколлинеарности в собственных данных. Изучить фактор инфляции дисперсии VIF.


```r
df = read.csv('COVID-19-in-Italy.csv')
df = df[df$region == 'Lombardia',]
```

```r
library(purrr)
num_df <- df %>% keep(is.numeric) #уберем стоблцы date и region
```

```r
library(corrplot)
col4 <- colorRampPalette(c("#7F0000", "red", "#FF7F00", "yellow",
                           "#7FFF7F", "cyan", "#007FFF", "blue","#00007F"))

corrplot(cor(num_df), method = "color", col = col4(20), cl.length = 21,
         order = "AOE", addCoef.col = "white")
```

Посмотрим на то, как связаны между собой показатели при предположении о том, что смертность зависит линейно от остальных показателей:

```r
colnames(num_df)
```

```r
library(car)
vif(lm(deaths ~ intensive_therapy + total_hospitalized + total_positive + total_cases,
       data = num_df))
```

Показатели intensive_therapy и total_hospitalized зависят от остальных почти линейно

#### 3. Изучить форму связи между переменными для выбранных данных.


Нелинейную связь обнаружим с помощью функции nlcor, а по графику примерно определим вид зависимости.

```r
library(nlcor)
```

```r
non_lin_corr = function (x, y, xlab, ylab) {
    plot(x, y, xlab = xlab, ylab = ylab)
    corr = nlcor(x, y)$adjusted.p.value
    print(corr)
    if (corr < 0.05) {
        'Correlation'
    } else {
        'No correlation'
    }
}
```

```r
non_lin_corr(num_df$deaths, num_df$total_cases, 'deaths', 'total_cases')
non_lin_corr(num_df$deaths, num_df$recovered, 'deaths', 'recovered')
non_lin_corr(num_df$deaths, num_df$new_positive, 'deaths', 'new_positive')
non_lin_corr(num_df$deaths, num_df$intensive_therapy, 'deaths', 'intensive_therapy')
non_lin_corr(num_df$deaths, num_df$total_hospitalized, 'deaths', 'total_hospitalized')
non_lin_corr(num_df$deaths, num_df$hospitalized_with_symptoms, 'deaths', 'hospitalized_with_symptoms')
non_lin_corr(num_df$deaths, num_df$total_positive, 'deaths', 'total_positive')
non_lin_corr(num_df$deaths, num_df$domestic_isolation, 'deaths', 'domestic_isolation')

non_lin_corr(num_df$total_cases, num_df$recovered, 'total_cases', 'recovered')
non_lin_corr(num_df$total_cases, num_df$new_positive, 'total_cases', 'new_positive')
non_lin_corr(num_df$total_cases, num_df$intensive_therapy, 'total_cases', 'intensive_therapy')
non_lin_corr(num_df$total_cases, num_df$total_hospitalized, 'total_cases', 'total_hospitalized')
non_lin_corr(num_df$total_cases, num_df$hospitalized_with_symptoms, 'total_cases', 'hospitalized_with_symptoms')
non_lin_corr(num_df$total_cases, num_df$total_positive, 'total_cases', 'total_positive')
non_lin_corr(num_df$total_cases, num_df$domestic_isolation, 'total_cases', 'domestic_isolation')

non_lin_corr(num_df$recovered, num_df$new_positive, 'recovered', 'new_positive')
non_lin_corr(num_df$recovered, num_df$intensive_therapy, 'recovered', 'intensive_therapy')
non_lin_corr(num_df$recovered, num_df$total_hospitalized, 'recovered', 'total_hospitalized')
non_lin_corr(num_df$recovered, num_df$hospitalized_with_symptoms, 'recovered', 'hospitalized_with_symptoms')
non_lin_corr(num_df$recovered, num_df$total_positive, 'recovered', 'total_positive')
non_lin_corr(num_df$recovered, num_df$domestic_isolation, 'recovered', 'domestic_isolation')

non_lin_corr(num_df$new_positive, num_df$intensive_therapy, 'new_positive', 'intensive_therapy')
non_lin_corr(num_df$new_positive, num_df$total_hospitalized, 'new_positive', 'total_hospitalized')
non_lin_corr(num_df$new_positive, num_df$hospitalized_with_symptoms, 'new_positive', 'hospitalized_with_symptoms')
non_lin_corr(num_df$new_positive, num_df$total_positive, 'new_positive', 'total_positive')
non_lin_corr(num_df$new_positive, num_df$domestic_isolation, 'new_positive', 'domestic_isolation')

non_lin_corr(num_df$intensive_therapy, num_df$total_hospitalized, 'intensive_therapy', 'total_hospitalized')
non_lin_corr(num_df$intensive_therapy, num_df$hospitalized_with_symptoms, 
             'intensive_therapy', 'hospitalized_with_symptoms')
non_lin_corr(num_df$intensive_therapy, num_df$total_positive, 'intensive_therapy', 'total_positive')
non_lin_corr(num_df$intensive_therapy, num_df$domestic_isolation, 'intensive_therapy', 'domestic_isolation')

non_lin_corr(num_df$total_hospitalized, num_df$hospitalized_with_symptoms, 
             'total_hospitalized', 'hospitalized_with_symptoms')
non_lin_corr(num_df$total_hospitalized, num_df$total_positive, 'total_hospitalized', 'total_positive')
non_lin_corr(num_df$total_hospitalized, num_df$domestic_isolation, 'total_hospitalized', 'domestic_isolation')

non_lin_corr(num_df$hospitalized_with_symptoms, num_df$total_positive, 
             'hospitalized_with_symptoms', 'total_positive')
non_lin_corr(num_df$hospitalized_with_symptoms, num_df$domestic_isolation, 
             'hospitalized_with_symptoms', 'domestic_isolation')

non_lin_corr(num_df$total_positive, num_df$domestic_isolation, 'total_positive', 'domestic_isolation')
```
