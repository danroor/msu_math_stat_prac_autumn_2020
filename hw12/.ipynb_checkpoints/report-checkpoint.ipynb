#### 1. Изучить метод наименьших квадратов.

МНК - метод, основанный на минимизации суммы квадратов отклонений некоторых функций от искомых переменных. Он может использоваться для «решения» переопределенных систем уравнений (когда количество уравнений превышает количество неизвестных), для поиска решения в случае обычных (не переопределенных) нелинейных систем уравнений, для аппроксимации точечных значений некоторой функции. МНК является одним из базовых методов регрессионного анализа для оценки неизвестных параметров регрессионных моделей по выборочным данным.

Пусть x — набор n неизвестных переменных, $ f_{i}(x),  i=1,\ldots ,m,  m>n$ — совокупность функций от этого набора переменных. Задача заключается в подборе таких значений x, чтобы значения этих функций были максимально близки к некоторым значениям $y_{i}$. По существу речь идет о «решении» переопределенной системы уравнений $ f_{i}(x)=y_{i},  i=1,\ldots ,m$ в указанном смысле максимальной близости левой и правой частей системы. Суть МНК заключается в выборе в качестве «меры близости» суммы квадратов отклонений левых и правых частей $ |f_{i}(x)-y_{i}|$. Таким образом, сущность МНК может быть выражена следующим образом:

$ \sum _{i}e_{i}^{2}=\sum _{i}(y_{i}-f_{i}(x))^{2}\rightarrow \min _{x}$

#### 2. Для собственных данных построить доверительные интервалы для модели регрессии и интервал предсказаний.

Рассмотрим линейную связь между количеством смертей и количеством случаев заболевания, выявленную в домашнем задании 10. Найдем доверительный интервал и интервал предсказаний для числа умерших при числе заболевших в 150 000 человек.

```r
df = read.csv('COVID-19-in-Italy.csv')
model = lm(deaths ~ total_cases, data = df)
```

```r
library(ggplot2)
ggplot(df, aes(x=total_cases, y=deaths)) + 
  geom_point(color='#2980B9', size = 4) + 
  geom_smooth(method=lm, color='#2C3E50')
```

```r
predict(model, newdata = data.frame(total_cases = 150000), interval = 'confidence')
```

```r
predict(model, newdata = data.frame(total_cases = 150000), interval = 'prediction')
```

#### 3. Модифицировать функцию regr() для расчета возраста Вселенной с помощью бутстрепа.

```r
library(boot)
library(gamair)
data(hubble)
```

```r
regr <- function(data) {
    
    res = boot( data = data,
                R = 1000,
                statistic = function(data, indices) {
    
                    dat <- data[indices, ]
                    fit <- lm(y ~ -1 + x, data = dat)
                    return(summary(fit)$coefficients[1])
                   
               } 
                   
    )
    
    q = quantile(res$t, c(0.025, 0.975), names = FALSE)
    q = c(q[2], q[1])
    
    result = 1 / (q * 60^2 * 24 * 365.25 / 3.09e19)
    
    names(result) = c('lower', 'upper')
    
    result
}

regr(hubble)
```
